```python
def P_MK(df):
    newdf = pd.DataFrame([[0]*20]*20, columns=[f'{i}' for i in range(20)], index=[f'{j}' for j in range(20)])
    for i in range(20):
      for j in range(20):
        newdf.iloc[i,j] = df.at['k',i]*df.at['M',j]
    return newdf
```

```python
Mk = P_MK(prob)
Mk
```

Це таблиця з такою ж індексацією, як і table (тобто рядки це k_i, ствопці це M_j), яка містить P(M_j, k_i) = P(M_j)*P(k_i)

Тепер необхідно знайти P(C) користуючись новою таблицею Mk та таблицею table

```python
def P_C(pmk, tbl, df):
    newrow = pd.Series([0]*len(df.columns), index=df.columns, name='C')
    for i in range(20):
      for j in range(20):
        newrow[int(tbl.iloc[i,j])] += pmk.iloc[i,j]
    df = df.append(newrow)
    return df
```

```python
prob = P_C(Mk, table, prob)
prob
```

Тепер необхідно знайти P(M, C), користуючись тими ж таблицями, але тепер сума не по парам, а лише по ключам

```python
def P_MC(pmk, tbl):
    newdf = pd.DataFrame([[0]*20]*20, columns=[f'{i}' for i in range(20)], index=[f'{j}' for j in range(20)])
    for i in range(20):
      for j in range(20):
        newdf.iloc[int(tbl.iloc[j,i]),i] += pmk.iloc[j,i]
    return newdf
```

```python
MC = P_MC(Mk, table)
MC
```

Маючи P(C) та P(M,C), можемо знайти P(M|C)

```python
def P_M_if_C(pr, pmc):
    newdf = pd.DataFrame([[0]*20]*20, columns=[f'{i}' for i in range(20)], index=[f'{j}' for j in range(20)])
    for i in range(20):
      for j in range(20):
        newdf.iloc[i, j] = round(pmc.iloc[i, j]/pr.at['C', i], 4)
    return newdf
```

```python
M_if_C = P_M_if_C(prob, MC)
M_if_C
```

Тепер необхідно зробити детермістичну вирішуючу функцію (ddf) та знайти її середні втрати, для цього нам необхідно просто реалізувати байєсівську вирішуючу функцію, адже вона і буде оптимальною



```python
def ddf(pm_if_c):
    indx_of_max = pm_if_c.idxmax(axis = 1)#виписали індекси тих елементів в кожному рядку де найбільше значення
    newdf = pd.DataFrame([[0]*20]*20, columns=[f'{i}' for i in range(20)], index=[f'{j}' for j in range(20)])
    for i in range(20):
      newdf.iloc[i, int(indx_of_max[i])] = 1
    return newdf
```

```python
indx_of_max = M_if_C.idxmax(axis = 1)
indx_of_max
```

```python
DDF = ddf(M_if_C)
DDF
```

Тут у вигляді таблиці представлено відповідність між шифротекстом та початковим відкритим текстом. Рядки це С, а стовпці це М

```python
def loss_ddf(pmc, ddf):
    loss = 0
    for i in range(20):
      for j in range(20):
        loss += pmc.iloc[i, j]*abs(1 - int(ddf.iloc[i, j]))
    return loss
```

```python
Loss_ddf = loss_ddf(MC, DDF)
print('loss for ddf = ', Loss_ddf)
```

Це ми знайшли середні втрати для двф

Тепер реалізуємо свф, навідміну від двф ми будемо враховувати всі максимальні значення, якщо їх декілка, для цього будемо знаходити власне максимальні значення, а не їх індекси, шукати їх та розподіляти в кожному рядку

```python
def sdf(pm_if_c, pr):
    max = pm_if_c.max(axis = 1)
    for i in range(20):
      Prmsum = 0
      for j in range(20):
        if pm_if_c.iloc[i, j] != max[i]:
          pm_if_c.iloc[i, j] = 0
        else:
          Prmsum += pr.at['M', j]
          pm_if_c.iloc[i, j] = pr.at['M', j]
      for j in range(20):
        pm_if_c.iloc[i, j] = pm_if_c.iloc[i, j]/Prmsum
    return pm_if_c
```

```python
SDF = sdf(M_if_C, prob)
SDF
```

Отримали таблицю розподілів свф. Тепер знайдемо середні втрати

```python
def loss_sdf(pmc, sdf):
    loss = 0
    for i in range(20):
      for j in range(20):
        loss += pmc.iloc[i, j]*abs(1 - int(sdf.iloc[i, j]))
    return loss
```

```python
Loss_sdf = loss_sdf(MC, SDF)
print('loss for sdf = ', Loss_sdf)
```
